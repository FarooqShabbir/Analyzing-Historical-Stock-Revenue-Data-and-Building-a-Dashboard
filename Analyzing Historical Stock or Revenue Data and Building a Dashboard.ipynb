```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
ticker = yf.Ticker("TSLA")
```

```python
tesla_data = ticker.history(period="max")
```

```python
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
response = requests.get(url)
html_data = response.text
```

```python
soup = BeautifulSoup(html_data, 'html.parser')
```

```python
tesla_revenue = pd.read_html(str(soup))[0]
```

```python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
```

```python
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
```

```python
print(tesla_revenue.tail())
```

```python
ticker_gme = yf.Ticker("GME")
```

```python
gme_data = ticker_gme.history(period="max")
```

```python
gme_data.reset_index(inplace=True)
print(gme_data.head())
```

```python
url_2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
response_2 = requests.get(url_2)
html_data_2 = response_2.text
```

```python
soup_2 = BeautifulSoup(html_data_2, 'html.parser')
```

```python
gme_revenue = pd.read_html(str(soup_2))[0]
gme_revenue["Revenue"] = gme_revenue['Revenue'].str.replace(',|\$',"", regex=True)
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]
```

```python
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt
def make_graph(stock_data, revenue_data, title):
    plt.figure(figsize=(14, 7))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Stock Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
make_graph(tesla_data, tesla_revenue, 'Tesla Stock Price')
```

```python
make_graph(gme_data, gme_revenue, 'GameStop Stock Price')
```
